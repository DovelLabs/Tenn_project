```python
book = openpyxl.load_workbook('Varsha-Transformation/Transformation/5-64.21UPWAV-TN-1-2022-10B.xlsx')

sheet = book.active
```

```python
State = sheet['A4'] # Column A and Row 4
Prior_Quater = sheet['A6']
Line = sheet['A7']
Waiver_Type = sheet['A8']
Waiver_Number = sheet['A9']
Waiver_Name= sheet['A10']
Type = sheet['A11']
Quarter_Ended = sheet['A5']
df = pd.read_excel("Varsha-Transformation/Transformation/5-64.21UPWAV-TN-1-2022-10B.xlsx",skiprows=11)
df = df.rename(columns={"Unnamed: 0": "Category", "Unnamed: 1": "Category Description"})
df.insert(7,"State", State.value, True)
df.insert(8, "Quarter_Ended", Quarter_Ended.value, True)
df.insert(9,"Prior_Quater", Prior_Quater.value, True)
df.insert(10,"Line",Line.value, True)
df.insert(11,"Waiver_Type",Waiver_Type.value, True)
df.insert(12,"Waiver_Number", Waiver_Number.value,True)
df.insert(13,"Waiver_Name",Waiver_Name.value, True)
df.insert(14,"Type",Type.value, True)
```

```python
df.to_csv('5.csv')
```
